```python
def get_lz_values(log):
    '''Get values of Lz operator for every orbital and translate them into labels of orbotals
        input:  lines    - contents of a file as a list of lines,
        output: orbitals - list of dicts containing MO indexes and labels (sigma, pi ...)'''

    labels = ["sigma", "pi", "delta", "phi", "gamma", "psi", "iota"]

    lz_patt = re.compile(r'\s*MO\s*(?P<nmo>\d+)\s*\(\s*(?P<shell>\d+)\s*\)\s*HAS LZ\(WEIGHT\)=\s*(?P<lz>-?\d*\.\d+)')
    
    res = list()
    
    for line in log.split('\n'):
        match = lz_patt.search(line)
        if match:
            nmo   = int(match.group('nmo'))
            shell = int(match.group('shell'))
            lz    = abs(int(float(match.group('lz'))))
            res.append({'index' : nmo-1, 'shell' : shell, 'lz' : lz})
    return res
            
```
