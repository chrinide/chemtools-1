```python
ccpv6z = bs.BasisSet.from_dict({"element" : "Be", "functions" : mp.parse_basis(ccpcv6z_mp)['Be']})
```

```python
import copy
def partial_wave_expand(basis):
    res = list()
    shells = basis.functions.keys()
    for i in range(1, len(shells)+1):
        bscopy = copy.copy(basis)
        bscopy.functions = {k:v for k, v in basis.functions.items() if k in shells[:i]}
        res.append(bscopy)
    return res
```

```python
lst = partial_wave_expand(ccpv6z)
for l in lst:
    print l.write_gamess()
```
